```python
import pandas as pd

# 1. Crear el DataFrame con 15 registros únicos (Nombres y datos originales)
data = {
    'nombre': ['Zaira', 'Yago', 'Vania', 'Ulises', 'Tania', 'Sacha', 'Raxel', 'Quara', 'Pavel', 'Oriel', 'Nidia', 'Mael', 'Lais', 'Kael', 'Jara'],
    'edad': [25, 19, 21, 24, 20, 22, 23, 18, 26, 21, 22, 20, 24, 25, 19],
    'ciudad': ['Cali', 'Medellín', 'Cali', 'Bogotá', 'Cartagena', 'Medellín', 'Cali', 'Bogotá', 'Cartagena', 'Cali', 'Pereira', 'Bogotá', 'Cali', 'Medellín', 'Pereira'],
    'nota_final': [91, 58, 74, 82, 63, 41, 89, 70, 52, 98, 45, 77, 85, 68, 93]
}

# Convertimos a DataFrame
df = pd.DataFrame(data)

# Mostramos el DataFrame original
print("--- DataFrame de Estudiantes ---")
print(df)
```

```python
# 2. Calcular estadísticas: promedio, máxima y mínima
promedio = df['nota_final'].mean()
nota_max = df['nota_final'].max()
nota_min = df['nota_final'].min()

print(f"\n--- Estadísticas de Notas ---")
print(f"Promedio de notas: {promedio:.2f}")
print(f"Nota máxima: {nota_max}")
print(f"Nota mínima: {nota_min}")
```

```python
# 3. Filtrar estudiantes con nota mayor o igual a 70
df_filtrado = df[df['nota_final'] >= 70]

print("\n--- Estudiantes con nota mayor o igual a 70 ---")
print(df_filtrado)
```

```python
# 4. Agrupar por ciudad y mostrar el promedio de notas por ciudad
promedio_por_ciudad = df.groupby('ciudad')['nota_final'].mean()

print("\n--- Promedio de notas por ciudad ---")
print(promedio_por_ciudad)
```

```python
# 5. Exportar el DataFrame filtrado a un archivo CSV
df_filtrado.to_csv('estudiantes_seleccionados.csv', index=False)

print("\n¡Proceso completado! El archivo 'estudiantes_seleccionados.csv' ha sido creado.")
```
