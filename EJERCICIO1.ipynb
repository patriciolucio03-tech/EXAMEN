```python
1# Definición de la lista de diccionarios con datos de ventas
ventas = [
    {"producto": "Laptop", "cantidad": 5, "precio_unitario": 800},
    {"producto": "Mouse", "cantidad": 15, "precio_unitario": 25},
    {"producto": "Teclado", "cantidad": 10, "precio_unitario": 45},
    {"producto": "Monitor", "cantidad": 7, "precio_unitario": 200}
]
```

```python
2# Inicializamos variables para los cálculos
ingreso_total_general = 0
producto_mas_vendido = ""
max_cantidad = 0

print("--- Total de ventas por producto ---")

for venta in ventas:
    # Calculamos el total por cada producto (cantidad * precio)
    total_producto = venta["cantidad"] * venta["precio_unitario"]
    print(f"Producto: {venta['producto']} | Total: ${total_producto}")

    # Sumamos al ingreso total de todas las ventas
    ingreso_total_general += total_producto

    # Lógica para identificar el producto más vendido (por cantidad)
    if venta["cantidad"] > max_cantidad:
        max_cantidad = venta["cantidad"]
        producto_mas_vendido = venta["producto"]

print("-" * 35)
```

```python
print(f"Producto más vendido: {producto_mas_vendido} ({max_cantidad} unidades)")
print(f"Ingreso total de todas las ventas: ${ingreso_total_general}")
```
