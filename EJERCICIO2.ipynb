```python
import random
# Generar una lista de 50 números aleatorios entre 1 y 100
numeros = [random.randint(1, 100) for _ in range(50)]
```

```python
# Inicializar listas para las categorías
bajo = []
medio = []
alto = []
```

```python
# Usar condicionales para filtrar los números
for num in numeros:
    if num < 33:
        bajo.append(num)
    elif 33 <= num <= 66:
        medio.append(num)
    else: # Mayores a 66
        alto.append(num)
```

```python
# Función para calcular promedio de forma segura
def calcular_promedio(lista):
    return sum(lista) / len(lista) if len(lista) > 0 else 0
```

```python
# Mostrar resultados: Conteo y Promedio
print(f"Resultados del filtrado:")
print("-" * 30)

print(f"Categoría 'Bajo'  (< 33):   Cantidad: {len(bajo)} | Promedio: {calcular_promedio(bajo):.2f}")
print(f"Categoría 'Medio' (33-66):  Cantidad: {len(medio)} | Promedio: {calcular_promedio(medio):.2f}")
print(f"Categoría 'Alto'  (> 66):   Cantidad: {len(alto)} | Promedio: {calcular_promedio(alto):.2f}")
```
